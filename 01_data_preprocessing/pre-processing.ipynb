```python
col = ['Location', 'Fuel_Type','Transmission',  'Owner_Type']
for i in col:
  print(df[i].unique())
```

```python
#import the required libraries for preprocessing
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler
```

```python
#one hot encoding
ohe1 = OneHotEncoder(sparse_output = False)
df1 = ohe1.fit_transform(df[['Location']])
df1 = pd.DataFrame(df1, columns = ohe1.get_feature_names_out(['Location']))

ohe2 = OneHotEncoder(sparse_output = False)
df2 = ohe2.fit_transform(df[['Fuel_Type']])
df2 = pd.DataFrame(df2, columns = ohe2.get_feature_names_out(['Fuel_Type']))

ohe3 = OneHotEncoder(sparse_output = False)
df3 = ohe3.fit_transform(df[['Transmission']])
df3 = pd.DataFrame(df3, columns = ohe3.get_feature_names_out(['Transmission']))

#label encoding
le = LabelEncoder()
df['Owner_Type'] = le.fit_transform(df['Owner_Type'])

df_final = pd.concat([df, df1, df2, df3], axis = 1)
df_final.drop(['Location', 'Fuel_Type', 'Transmission'], axis = 1, inplace = True)
```

```python
df_final.info()
```

```python
data = ['Engine', 'Power', 'Mileage']
for i in data:
  print(df_final[i].unique())
  print('*'*100)
```

```python
# Cleaning 'Engine' column
df_final['Engine'] = df_final['Engine'].str.replace('CC', '')
df_final['Engine'] = df_final['Engine'].astype(float)

# Cleaning 'Power' column
df_final['Power'] = df_final['Power'].str.replace('bhp', '')
df_final['Power'] = df_final['Power'].replace('null ', np.nan)
df_final['Power'] = df_final['Power'].bfill()
df_final['Power'] = df_final['Power'].astype(float)

# Cleaning 'Mileage' column
def milege_conversion (value):
    if 'km/kg' in value :
        return float(value.replace('km/kg','')) * 1.5
    
    elif 'kmpl' in value :
        return float(value.replace('kmpl',''))
    
    else:
        return float(value)
    
df_final['Mileage'] = df_final['Mileage'].apply(milege_conversion)
```
