```python
"""
Note: Lasso (Least Absolute Shrinkage and Selection Operator) adds a penalty 
equal to the absolute value of the magnitude of coefficients. Unlike Ridge, Lasso 
can shrink some coefficients exactly to zero, making it highly useful for automatic 
feature selection in datasets with many irrelevant variables.
"""
```

Import Libraries

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

Load Dataset

```python
from sklearn.datasets import fetch_california_housing
```

Import Lasso Regression Model

```python
from sklearn.linear_model import Lasso
```

Feature Selecton

```python
housing = fetch_california_housing()
X, y = housing.data, housing.target
```

Train-Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

Feature Scaling

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

Model training and prediction

```python
model = Lasso(alpha = 0.01)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

Model evaluation

```python
print(f"Regression MAE: {mean_absolute_error(y_test, y_pred):.4f}\n")
print(f"Regression MSE: {mean_squared_error(y_test, y_pred):.4f}\n")
print(f"Regression RMSE: {np.sqrt(mean_squared_error(y_test, y_pred)):.4f}\n")
```

```python
train_accuracy = r2_score(y_train, model.predict(X_train))
test_accuracy = r2_score(y_test, y_pred)
gap = abs(train_accuracy - test_accuracy)

if train_accuracy > test_accuracy + 0.1:
    print('overfit')
elif test_accuracy < 0.6:
    print('underfit')
else:
    print('goodfit')
```
