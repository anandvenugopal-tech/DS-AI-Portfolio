Import libraries

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score, confusion_matrix, mean_absolute_error
import matplotlib.pyplot as plt
import seaborn as sns
```

Load Dataset

```python
from sklearn.datasets import fetch_california_housing
```

Import Linear Regresson Model

```python
from sklearn.linear_model import LinearRegression
```

Feature Selection

```python
housing = fetch_california_housing()
X, y = housing.data, housing.target
```

Train-Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

Feature Scaling

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

Model training and prediction

```python
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

Model Evaluation

```python
print(f"Regression MAE: {mean_absolute_error(y_test, y_pred):.4f}\n")
print(f"Regression MSE: {mean_squared_error(y_test, y_pred):.4f}\n")
print(f"Regression RMSE: {np.sqrt(mean_squared_error(y_test, y_pred)):.4f}\n")
```

```python
train_accuracy = r2_score(y_train, model.predict(X_train))
test_accuracy = r2_score(y_test, y_pred)
gap = abs(train_accuracy - test_accuracy)

if gap <= 0.2:
    print('Good fit')
elif gap > 0.2:
    print('Overfit')
else:
    print('Underfit')
```
