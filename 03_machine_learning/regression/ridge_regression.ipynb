Import Libraries

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score, confusion_matrix, mean_absolute_error
```

Load Dataset

```python
from sklearn.datasets import fetch_california_housing
```

Import Ridge Regression Model

```python
from sklearn.linear_model import Ridge
```

Feature Selecton

```python
housing = fetch_california_housing()
X, y = housing.data, housing.target
```

Train-Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

Model training and prediction

```python
model = Ridge(alpha = 0.01)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

Model evaluation

```python
print(f"Regression MAE: {mean_absolute_error(y_test, y_pred):.4f}\n")
print(f"Regression MSE: {mean_squared_error(y_test, y_pred):.4f}\n")
print(f"Regression RMSE: {np.sqrt(mean_squared_error(y_test, y_pred)):.4f}\n")
```

```python
train_accuracy = r2_score(y_train, model.predict(X_train))
test_accuracy = r2_score(y_test, y_pred)

if train_accuracy > test_accuracy + 0.1:
    print('overfit')
elif test_accuracy < 0.6:
    print('underfit')
else:
    print('goodfit')
```

```python
print(f'Train Accuaracy: {train_accuracy}')
print(f'Test Accuracy: {test_accuracy}')
```
