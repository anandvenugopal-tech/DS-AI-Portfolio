```python
import cv2 as cv
```

```python
cap = cv.VideoCapture('car-detection.mp4')

if not cap.isOpened():
    print("❌ Error: Video file not opened")
    exit()
```

```python
car_cascade = cv.CascadeClassifier('cars.xml')
```

```python
while True:
    ret, frame = cap.read()
    if not ret:
        print("✅ End of video OR frame not read")
        break
    gray = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)
    cars = car_cascade.detectMultiScale(gray,1.1, 5)
    for (x, y, w, h) in cars:
        cv.rectangle(frame, (x, y), (x+w, y+h), (0, 0, 255), 5)
    
    cv.imshow('Car detection', frame)
    if cv.waitKey(30) == ord('q'):
        break

cap.release()
cv.destroyAllWindows()
```
